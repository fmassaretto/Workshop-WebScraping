# Pandas DataFrames
DataFrames no Pandas são uma estrutura de dados bidimensional (2-D) definida no pandas que consiste em linhas e colunas.



```python
import pandas as pd
```

**Criando um DataFrame**

Criação de data frame com os valores das series da oficina passada.

```python
numeric_data = [1, 7, 2] # dado numérico

index = ["x", "y", "z"] # índices pré-definido

data = pd.Series(numeric_data, index)

df = pd.DataFrame(data)

print(df)
```

Aqui, data pode ser:


1.   Um ou mais dicionários
2.   uma ou mais Series
3.   2D-numpy Ndarray (Array multidimensional)

Lembrando que index por padrão começa por 0: 0, 1, 2 ..., (n-1)



**Criar um DataFrame do Pandas a partir de vários dicionários**

Criando um Dataframe com dois dicionários.

```python
# Definição dicionario 1
dict1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4}

# Definição dicionario 2
dict2 = {'a': 5, 'b': 6, 'c': 7, 'd': 8, 'e': 9}

# Definição de Data com dict1 e dict2
Data = {'primeiro': dict1, 'segundo': dict2}

# Create DataFrame
df = pd.DataFrame(Data)

print(df)
```

**Converter lista de dicionários em um DataFrame do Pandas**

Aqui, pegamos três dicionários e com a ajuda da função from_dict() os convertemos em Pandas DataFrame.

```python
data_c = [
  {'A': 5, 'B': 0, 'C': 3, 'D': 3},
  {'A': 7, 'B': 9, 'C': 3, 'D': 5},
  {'A': 2, 'B': 4, 'C': 7, 'D': 6, 'E': 32}
 ]

df = pd.DataFrame.from_dict(data_c, orient='columns')

print(df)
```

**Criar DataFrame de várias séries**

Criando um dataframe de três séries.

```python
# Definição da series 1
s1 = pd.Series([1, 3, 4, 5, 6, 2, 9])

# Definição da series 2
s2 = pd.Series([1.1, 3.5, 4.7, 5.8, 2.9, 9.3])

# Definição da series 3
s3 = pd.Series(['a', 'b', 'c', 'd', 'e'])

# Definição de Data
Data = {'primeiro':s1, 'segundo':s2, 'terceiro':s3}

# Criar DataFrame
df = pd.DataFrame(Data)

print(df)
```

**Converter um array em Pandas Dataframe**

Ao criar um DataFrame de arrays 2D, as dimensões do array 2D devem ser as mesmas.

```python
# Definição de array 2d 1
d1 =[[2, 3, 4], [5, 6, 7], [2,3,5]]

# Definição de array 2d 2
d2 =[[2, 4, 8], [1, 3, 9], [2.23343,'B',0]]

# Definição de Data
Data ={'primeiro': d1, 'segundo': d2}

# Create DataFrame
df_2d = pd.DataFrame(Data)

df_2d
```

# Acesso e Manipulação dos dados

Lendo dados de um arquivo do tipo .csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/fmassaretto/Workshop-WebScraping/main/arquivos/employees.csv")

# print(type(df))

# print(df)

# df.head(30)
df.tail(10)
```

**isnull**

Conferindo quais valores sao nulos, retorna true se o valor for nulo e false se tem algum valor na celula

```python
df.isnull()
```

**fillna**

Preeenchendo um valor que você definir no lugar dos valores nulos (no caso do pandas, ele substitui nulo (None) por NaN (Not a Number))

```python

# df["Gender"].fillna("sem valor") # substirui um campo nulo (NaN) de uma coluna específica
df.fillna("nao ok", inplace = True) # substirui um campo nulo (NaN) de todo o dataframe
df.tail(50)
```

**dropna**

Remove linhas que contenham valor ou valores nulos

```python
df.dropna() # remove qualquer linha que tenha pelo menos um valor nulo
```

```python
data = df.dropna(how = 'all') # remove a linha em que todos os valores são nulos
data
```

**loc e iloc**

loc e iloc são utilizados para fatiar (separar) um conjunto de dados de um dataframe.

Com loc, conseguimos acessar multiplas colunas e fazer alguma logica que só retorne os dados que queremos analisar.

```python
df.loc[df['Gender'] == 'Male']
```

```python
df.loc[(df['Gender'] == 'Male') & (df['Salary'] > 100_000.0)].head()
```

```python
df.loc[0:10, ['First Name','Salary']]
```

iloc é baseado na posição onde quero encontrar o dado, ex.:

```python
df.iloc[0, 4]
```

sendo,

primeiro parametro: linha=0

segundo parametro: coluna=4

Também posso ter uma faixa de dados, ex.:

```python
df.iloc[0:5, 3:6]
```
